```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.optimize as opt
df = pd.read_csv('ex2data2.txt',delimiter = ',',names=['SCORE1','SCORE2','PASSED'])
x = df.values[:,0:2]
y = df.values[:,-1]
y = y.reshape((y.shape[0],1))
```

```python
pos = np.array([x[i] for i in range(x.shape[0]) if y[i]==1])
neg = np.array([x[i] for i in range(x.shape[0]) if y[i]==0])
plt.scatter(pos[:,0],pos[:,1],marker = 'o',color='g',label='PASSED')
plt.scatter(neg[:,0],neg[:,1],marker = 'x',color='r',label='REJECTED')
plt.xlabel("MICROCHIP TEST 1")
plt.ylabel('MICROCHIP TEST 2')
plt.legend()
plt.show()
```

```python
def addpoly(x1,x2):
    degree = 6
    out = np.ones((x.shape[0]))[:,np.newaxis]
    for i in range(1,degree+1):
        for j in range(i+1):
            out = np.hstack((out,(np.multiply(np.power(x1,i-j),np.power(x2,j)))[:,np.newaxis]))
    return out
x= addpoly(x[:,0], x[:,1])
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
def cost(theta_t,X_t,y_t,lam_t):
    m = len(y_t)
    J = (-1/m) * (y_t.T @ np.log(sigmoid(X_t @ theta_t)) + (1 - y_t.T) @ np.log(1 - sigmoid(X_t @ theta_t)))
    reg = (lam_t/(2*m)) * (theta_t[1:].T @ theta_t[1:])
    J = J + reg
    return J
```

```python
def lrGradientDescent(theta, X, y, lambda_t):
    m = len(y)
    grad = np.zeros([m,1])
    grad = (1/m) * X.T @ (sigmoid(X @ theta) - y)
    grad[1:] = grad[1:] + (lambda_t / m) * theta[1:]
    return grad
```

```python
lam=1
theta = np.zeros((x.shape[1],1))
J = cost(theta,x, y, lam)
g = (lrGradientDescent(theta,x,y,lam))
```

```python
output = opt.fmin_tnc(func = cost, x0 = theta.flatten(), fprime = lrGradientDescent,args = (x, y.flatten(), lam))
theta = output[0]
```

```python
u = np.linspace(-1, 1.5, 50)
v = np.linspace(-1, 1.5, 50)
z = np.zeros((len(u), len(v)))
def mapFeatureForPlotting(X1, X2):
    degree = 6
    out = np.ones(1)
    for i in range(1, degree+1):
        for j in range(i+1):
            out = np.hstack((out, np.multiply(np.power(X1, i-j), np.power(X2, j))))
    return out
for i in range(len(u)):
    for j in range(len(v)):
        z[i,j] = np.dot(mapFeatureForPlotting(u[i], v[j]), theta)
mask = y.flatten() == 1
plt.scatter(pos[:,0],pos[:,1],marker = 'o',color='g',label='PASSED')
plt.scatter(neg[:,0],neg[:,1],marker = 'x',color='r',label='REJECTED')
plt.contour(u,v,z,0)
plt.xlabel('Microchip Test1')
plt.ylabel('Microchip Test2')
plt.legend()
plt.show()
```

```python
def accuracy(X, y, theta, cutoff):
    pred = [sigmoid(np.dot(X, theta)) >= cutoff]
    acc = np.mean(pred==y)
    print(acc * 100)
accuracy(x, y.flatten(), theta, 0.5)
```
